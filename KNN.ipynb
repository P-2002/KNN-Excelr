```python
Q1 = df['legs'].quantile(0.25)
Q3 = df['legs'].quantile(0.75)
IQR = Q3 - Q1
```

```python
from sklearn.preprocessing import StandardScaler
X = df.drop(columns=['animal name'])
y = df['type'].astype(int)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(pd.DataFrame(X_scaled, columns=X.columns).head())
```

**3. Train and Test Data**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

**4 & 5. Implement KNN and Distance Metric with Best K**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2) #Euclidean
knn.fit(X_train, y_train)
```

**6. Evaluate Classifier Performance**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

y_pred = knn.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))
print("Precision (macro):",precision_score(y_test, y_pred, average='macro'))
print("Recall (macro):",recall_score(y_test, y_pred, average='macro'))
print("F1 Score (macro):",f1_score(y_test, y_pred, average='macro'))
print("\nClassification Report:\n",classification_report(y_test, y_pred))
```

**7. Decision Boundaries**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

knn_pca = KNeighborsClassifier(n_neighbors=3, metric='minkowski', p=2)
knn_pca.fit(X_pca, y)

x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                     np.linspace(y_min, y_max, 300))
Z = knn_pca.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4, cmap='tab10')
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolors='k', cmap='tab10')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Decision Boundaries with PCA')
plt.show()
```

**Interview Questions:**


1.  What are the key hyperparameters in KNN?

**Answer:**

*   n_neighbors - Number of nearest neighbors to consider for making predictions.

*  metric - The distance function used to compute similarity between points.

*   weights - How to weight the votes of neighbors.

* p(only if metric='minkowski') - Power parameter for minkowski distance.

* algorithm - the algorithm used to compute nearest neighbor.

2.   What distance metrics can be used in KNN?

**Answer:**

* Euclidean - Best for continuous numeric data.

* Manhattan - Suitable for high-dimensional/sparse data.

* Minkowski - Generalized form of Euclidean and Manhattan.

* Chebyshev - Takes the minimum absolute difference along any feature.

* Hamming - For categorical and binary data.

* Cosine - Measures angle between vectors.
